# Manipulación

```python
# Filtramos para utilizar estas columnas en nuestro analisis
filtroencora = encora[['Team_Velocity', 'Active_team_members', 'Team']]
filtrointime = inTime[['Team_Velocity', 'Active_team_members', 'Team']]
filtrocii = Cii[['Team_Velocity', 'Active_team_members', 'Team']]
filtrocra = cra[['Title','SES_Doc_delivered_on_time','Release_Notes_delivered_on_time','TypeProyect']]
filtrocodebase = codebase[['Title','SES_Doc_delivered_on_time','Release_Notes_delivered_on_time','TypeProyect']]
```

```python
# Valida el tipo de dato
filtroencora.info()
filtrointime.info()
filtrocii.info()
filtrocra.info()
filtrocodebase.info()
```

```python
# reseteo el index para poder tomar los valores de la fecha
filtroencora = filtroencora.reset_index().sort_index()
# Le agrego el formato de fecha correspondiente
filtroencora['Sprint_Date'] = pd.to_datetime(filtroencora['Sprint_Date'],format='mixed')

filtrointime = filtrointime.reset_index().sort_index()
filtrointime['Sprint_Date'] = pd.to_datetime(filtrointime['Sprint_Date'],format='mixed')

filtrocii = filtrocii.reset_index().sort_index()
filtrocii['Sprint_Date'] = pd.to_datetime(filtrocii['Sprint_Date'],format='mixed')

filtrocra = filtrocra.reset_index().sort_index()
filtrocra['Date'] = pd.to_datetime(filtrocra['Date'],format='mixed')

filtrocodebase = filtrocodebase.reset_index().sort_index()
filtrocodebase['Date'] = pd.to_datetime(filtrocodebase['Date'],format='mixed')

#Agrego el nombre del Mes
filtroencora['Mes'] = (filtroencora['Sprint_Date']).dt.month_name()
filtrointime['Mes'] = (filtrointime['Sprint_Date']).dt.month_name()
filtrocii['Mes'] = (filtrocii['Sprint_Date']).dt.month_name()
filtrocra['Mes'] = (filtrocra['Date']).dt.month_name()
filtrocodebase['Mes'] = (filtrocodebase['Date']).dt.month_name()

#Agrego el año (Anio)
filtroencora['Anio'] = (filtroencora['Sprint_Date']).dt.year
filtrointime['Anio'] = (filtrointime['Sprint_Date']).dt.year
filtrocii['Anio'] = (filtrocii['Sprint_Date']).dt.year
filtrocra['Anio'] = (filtrocra['Date']).dt.year
filtrocodebase['Anio'] = (filtrocodebase['Date']).dt.year

#Agrego el mes
filtroencora['Nro Mes'] = (filtroencora['Sprint_Date']).dt.month
filtrointime['Nro Mes'] = (filtrointime['Sprint_Date']).dt.month
filtrocii['Nro Mes'] = (filtrocii['Sprint_Date']).dt.month
filtrocra['Nro Mes'] = (filtrocra['Date']).dt.month
filtrocodebase['Nro Mes'] = (filtrocodebase['Date']).dt.month

filtroencora.head()
filtrointime.head()
filtrocii.head()
filtrocra.head()
filtrocodebase.head()
```

# Visualización y Conclusiónes

```python
# Se importa la libreria matplotlib para crear los graficos
import matplotlib.pyplot as plt

# Convertir Sprint_Date a datetime porque si no no me funciona para agregar la fecha como eje x
agrupacion1['Sprint_Date'] = pd.to_datetime(agrupacion1['Sprint_Date'])

# Preparar los datos para la creacion del gráfico
# creo una variable teams con los 3 grupos para poder recorrerlo y mostrarlo
teams = agrupacion1['Team'].unique()
# Hago un vector con las fechas para poder utilizarlo como indice y le agrego formato de fecha universal, le agrego el .unique para quitar duplicados
dates = agrupacion1['Sprint_Date'].dt.strftime('%Y-%m-%d').unique()
x = np.arange(len(dates))
# agrego un ancho a las barras
width = 0.25

# Crear el gráfico
plt.figure(figsize=(12, 6))

# hago un for para añadir una barra para cada team
# agrego el enumerate() para poder tomar el indice de cada equipo y poder dibujar correctamente las barras
# i = indice del equipo
for i, team in enumerate(teams):
    team_data = agrupacion1[agrupacion1['Team'] == team]
    bars = plt.bar(
        x + i * width,  # Muestra las barras para cada equipo
        team_data['Team_Velocity'], # Altura de las barras, que en este caso es el team_velocity de cada equipo
        width=width, # el ancho de las barras que declaramos arriba
        label=team + ' Team Velocity' # etiqueta del equipo
    )
    # Hago el for para agregar las etiquetas de Team Velocity en las barras
    for bar, velocity in zip(bars, team_data['Team_Velocity']):
        plt.text(
            bar.get_x() + bar.get_width() / 2,  # X: centro de la barra
            bar.get_height() + 1,  # Y: Esto coloca el texto un poco por encima de la altura de la barra (+1), para que el texto no quede sobrepuesto con la barra misma.
            f'{int(velocity)}',  # Mostrar el valor de Team Velocity
            ha='center', # Horizontal alignment
            va='bottom', # Vertical alignment
            fontsize=10,
            color='black'
        )

# Personalizar el gráfico
plt.xlabel('Fecha')
plt.ylabel('')
plt.title('Team Velocity')
plt.xticks(x + width, dates)  # Alinear las etiquetas de las fechas
plt.legend(title='Team')
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

### Conclusión del grafico anterior: En el gráfico anterior se puede observar el desempeño de los diferentes equipos en los últimos 9 sprints, esta métrica es utilizada para evaluar la velocidad con la que los 3 equipos operan basados en story points.

```python
# Agrupar por 'TypeProyect' y contar los valores de 'Release_Notes_delivered_on_time', fill_value=0 -> esto es porque algunos de los valores que colocamos como eje x no tiene datos entonces le asignamos el 0
agrupacion = DataCompare2.groupby(['TypeProyect', 'Release_Notes_delivered_on_time'])['Release_Notes_delivered_on_time'].count().unstack(fill_value=0)

# Crear subgráficos
fig, axes = plt.subplots(nrows=1, ncols=len(agrupacion.index), figsize=(15, 6), sharey=True)

# Iterar sobre cada proyecto para crear gráficos individuales
for i, (project, data) in enumerate(agrupacion.iterrows()):
    ax = axes[i]
    data.plot(kind='bar', ax=ax, color='#3a4e7a')

    # Personalizar cada subgráfico
    ax.set_title(f'{project} Release Notes delivered on time', fontsize=12)
    ax.set_xlabel('', fontsize=10)
    ax.set_ylabel('', fontsize=10 if i == 0 else 0)  # Solo el primer gráfico muestra el eje Y
    ax.tick_params(axis='x', rotation=45)

    # Agregar etiquetas encima de las barras
    for j, value in enumerate(data.values):
        ax.text(j, value + 0.1, str(value), ha='center', va='bottom', fontsize=9)

# Ajustar diseño
plt.tight_layout() #Matplotlib ajusta automáticamente los márgenes y espacios entre los gráficos. tambien se puede usar constrained_layout
plt.show()
```

### Conclusión del grafico anterior: En el gráfico anterior se puede observar la métrica acerca de los release notes para CRA y Codebase, la cúal tiene el propósito de medir si se esta haciendo entrega de dichos documentos en las ventanas de tiempo establecidas, que son 12 horas antes de realizar el deploy a producción.
