# Data manipulation

```python
# We filter to use these columns in our analysis
filtroA = A[['Team_Velocity', 'Active_team_members', 'Team']]
filtroB = B[['Team_Velocity', 'Active_team_members', 'Team']]
filtroC = C[['Team_Velocity', 'Active_team_members', 'Team']]
filtroProject1 = Project1[['Title','SES_Doc_delivered_on_time','Release_Notes_delivered_on_time','TypeProyect']]
filtroProject2 = Project2[['Title','SES_Doc_delivered_on_time','Release_Notes_delivered_on_time','TypeProyect']]
```

```python
# Validates the data type
filtroA.info()
filtroB.info()
filtroC.info()
filtroProject1.info()
filtroProject2.info()
```

```python
# Reset the index to be able to take the date values
filtroA = filtroA.reset_index().sort_index()
# Add the corresponding date format
filtroA['Sprint_Date'] = pd.to_datetime(filtroA['Sprint_Date'],format='mixed')

filtroB = filtroB.reset_index().sort_index()
filtroB['Sprint_Date'] = pd.to_datetime(filtroB['Sprint_Date'],format='mixed')

filtroC = filtroC.reset_index().sort_index()
filtroC['Sprint_Date'] = pd.to_datetime(filtroC['Sprint_Date'],format='mixed')

filtroProject1 = filtroProject1.reset_index().sort_index()
filtroProject1['Date'] = pd.to_datetime(filtroProject1['Date'],format='mixed')

filtroProject2 = filtroProject2.reset_index().sort_index()
filtroProject2['Date'] = pd.to_datetime(filtroProject2['Date'],format='mixed')

# Add the name of the Month
filtroA['Mes'] = (filtroA['Sprint_Date']).dt.month_name()
filtroB['Mes'] = (filtroB['Sprint_Date']).dt.month_name()
filtroC['Mes'] = (filtroC['Sprint_Date']).dt.month_name()
filtroProject1['Mes'] = (filtroProject1['Date']).dt.month_name()
filtroProject2['Mes'] = (filtroProject2['Date']).dt.month_name()

# Add the year (Anio)
filtroA['Anio'] = (filtroA['Sprint_Date']).dt.year
filtroB['Anio'] = (filtroB['Sprint_Date']).dt.year
filtroC['Anio'] = (filtroC['Sprint_Date']).dt.year
filtroProject1['Anio'] = (filtroProject1['Date']).dt.year
filtroProject2['Anio'] = (filtroProject2['Date']).dt.year

# Add the month
filtroA['Nro Mes'] = (filtroA['Sprint_Date']).dt.month
filtroB['Nro Mes'] = (filtroB['Sprint_Date']).dt.month
filtroC['Nro Mes'] = (filtroC['Sprint_Date']).dt.month
filtroProject1['Nro Mes'] = (filtroProject1['Date']).dt.month
filtroProject2['Nro Mes'] = (filtroProject2['Date']).dt.month

filtroA.head()
filtroB.head()
filtroC.head()
filtroProject1.head()
filtroProject2.head()
```

```python
# The files are joined, since they have the same columns, a concat is performed
DataCompare = pd.concat([filtroA,filtroB,filtroC], ignore_index=True, axis = 0)
DataCompare2 = pd.concat([filtroProject1,filtroProject2], ignore_index=True, axis = 0)
# The data is displayed
DataCompare.head()
```

# Visualization and Conclusions

```python
# Import the matplotlib library to create the graphics
import matplotlib.pyplot as plt

# Convert Sprint_Date to datetime because otherwise it won't work for me to add the date as an x-axis
agrupacion1['Sprint_Date'] = pd.to_datetime(agrupacion1['Sprint_Date'])

# Prepare the data for creating the chart
# Create a variable "teams" with the 3 groups so I can iterate through it and display it
teams = agrupacion1['Team'].unique()
# I make a vector with the dates to be able to use it as an index and I add a universal date format, I add the .unique to remove duplicates
dates = agrupacion1['Sprint_Date'].dt.strftime('%Y-%m-%d').unique()
x = np.arange(len(dates))
# add a width to the bars
width = 0.25

# Create the chart
plt.figure(figsize=(12, 6))

# I make a for to add a bar for each team
# I add the enumerate() to be able to take the index of each team and be able to draw the bars correctly
# i = team index
for i, team in enumerate(teams):
    team_data = agrupacion1[agrupacion1['Team'] == team]
    bars = plt.bar(
        x + i * width,  # Displays bars for each team
        team_data['Team_Velocity'], # Height of the bars, which in this case is the team_velocity of each team
        width=width, # the width of the bars we declared above
        label=team + ' Team Velocity' # team tag
    )
    # I do the for to add the Team Velocity labels to the bars
    for bar, velocity in zip(bars, team_data['Team_Velocity']):
        plt.text(
            bar.get_x() + bar.get_width() / 2,  # X: center of the bar
            bar.get_height() + 1,  # Y: This places the text slightly above the height of the bar (+1), so that the text does not overlap the bar itself.
            f'{int(velocity)}',  # Display the Team Velocity value
            ha='center', # Horizontal alignment
            va='bottom', # Vertical alignment
            fontsize=10,
            color='black'
        )

# Customize the chart
plt.xlabel('Fecha')
plt.ylabel('')
plt.title('Team Velocity')
plt.xticks(x + width, dates)  # Align date labels
plt.legend(title='Team')
plt.tight_layout()

# Show the graph
plt.show()
```

### Conclusion from the previous graph: In the previous graph you can see the performance of the different teams in the last 9 sprints, this metric is used to evaluate the speed with which the 3 teams operate based on story points.

```python
# Group by 'TypeProyect' and count the values of 'Release_Notes_delivered_on_time', fill_value=0 -> this is because some of the values we placed as the x axis have no data so we assign them 0
agrupacion = DataCompare2.groupby(['TypeProyect', 'Release_Notes_delivered_on_time'])['Release_Notes_delivered_on_time'].count().unstack(fill_value=0)

# Create subgraphs
fig, axes = plt.subplots(nrows=1, ncols=len(agrupacion.index), figsize=(15, 6), sharey=True)

# Iterate over each project to create individual charts
for i, (project, data) in enumerate(agrupacion.iterrows()):
    ax = axes[i]
    data.plot(kind='bar', ax=ax, color='#3a4e7a')

    # Customize each subgraph
    ax.set_title(f'{project} Release Notes delivered on time', fontsize=12)
    ax.set_xlabel('', fontsize=10)
    ax.set_ylabel('', fontsize=10 if i == 0 else 0)  # Only the first chart shows the Y axis
    ax.tick_params(axis='x', rotation=45)

    # Add labels above the bars
    for j, value in enumerate(data.values):
        ax.text(j, value + 0.1, str(value), ha='center', va='bottom', fontsize=9)

# Adjust layout
plt.tight_layout() #M atplotlib automatically adjusts margins and spaces between plots. You can also use constrained_layout
plt.show()
```

### Conclusion from the previous graph: In the previous graph you can see the metrics about the release notes for Project1 and Project2, which aims to measure whether these documents are being delivered within the established time windows, which are 12 hours before deploying to production.
